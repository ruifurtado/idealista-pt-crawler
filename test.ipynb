```python
import undetected_chromedriver as uc
from selenium import webdriver
import pandas as pd


# url
#url = "https://www.idealista.pt/comprar-casas/lisboa-distrito/"
url ="https://www.idealista.pt/comprar-casas/lisboa-distrito/com-preco-max_300000,tamanho-min_80,t2,t3,t4-t5/"
df = pd.DataFrame(columns=["title","link","topology","area","price","garage","house","description"])
```

```python
import re
import time
from random import randrange, uniform
import undetected_chromedriver as uc
from selenium import webdriver
from bs4 import BeautifulSoup
import traceback

url ="https://www.idealista.pt/comprar-casas/lisboa-distrito/com-preco-max_300000,tamanho-min_80,t2,t3,t4-t5/"
page = url
print(page)
count=1
for x in range(0,5000): # maybe change this idealista can only retrieve 60 pages
    try:
        ## Example Proxy
        options = webdriver.ChromeOptions() 

        options.add_argument("start-maximized")
        #options.add_argument(f'--proxy-server={PROXY}')

        driver = uc.Chrome(options=options)
        
        try:
            driver.get(page)
        except Exception as e:
            # Close browser
            driver.close()
            driver.quit()
            print('Error! We were caught!',e)
            n = 0 
            while n<10:
                time.sleep(uniform(5,10))
                try:
                    driver.get(page)
                    break
                except:
                    print('Didnt work!! Try nr:',n)
        
        elem = driver.find_element("xpath", "//*")
        source_code = elem.get_attribute("outerHTML")
        soup = BeautifulSoup(source_code, 'html.parser')
    
        soup.find_all('div', attrs={"class": "price-row"})
        
        #print(page)
        #<div class="item-info-container">

        for house in soup.find_all('div', attrs={"class": "item-info-container"}):
            title = house.find_all('a', attrs={"aria-level": 2})[0]['title']
            link =  house.find_all('a', attrs={"aria-level": 2})[0]['href']
            #print('Title: ',title)
            #print('Link: ', house.find_all('a')[1]['href'])
            price=re.sub(r"\s+", "", house.find('span', attrs={"class": "item-price h2-simulated"}).get_text().replace("\\n", "").strip())
            try:
                garage=house.find('span', attrs={"class": "item-parking"}).get_text().replace("\\n", "").strip()
                flat=house.find_all('span', attrs={"class": "item-detail"})[2].get_text().replace("\\n", "").strip()
                house_type="prédio"
            except:
                garage=""
                flat=""
                house_type="moradia"
            topology=house.find_all('span', attrs={"class": "item-detail"})[0].get_text().replace("\\n", "").strip()
            area=house.find_all('span', attrs={"class": "item-detail"})[1].get_text().replace("\\n", "").strip()
            description = house.find_all('p', attrs={"class": "ellipsis"})[0].get_text().replace("\\n", "").strip()
            #print("Topology:", topology)
            #print("Area:", area)
            #print("Flat:",flat)
            #print('Price:', price)
            #print('Garage:', garage)
            #print('House:',house_type)
            #print('Description:',description)
            df.loc[len(df.index)] = [
                title,
                link,
                topology,
                area,
                price,
                garage,
                house_type,
                description
            ]
        
        # Close browser
        driver.close()
        driver.quit()
        
    except Exception as e:
        print('error',e)
        traceback.print_exc()
        break
    time.sleep(uniform(2, 3.5))
    count+=1  
    page = url+"pagina-"+str(count)
    print('\n')
    print(page)
```

```python
df.to_csv('houses_20231003.csv',header=False)
```
